折线图

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams


#画布
plt.figure(figsize = (10, 5))
month = ['1月', '2月', '3月', '4月']
sales = [100, 200, 150, 180]

#绘制折线图
plt.plot(month, sales)

#显示图
plt.show()
```

```python
#后面放字体库C:\Windows\Fonts这里有
#mac电脑是 'STHeiti'
#HP Simplified Hans
rcParams['font.family'] = 'SimHei'

plt.plot(month, sales)
plt.show()
```

```python
plt.title('销售', color = 'pink', fontsize = 20)
plt.xlabel('月份', fontsize = 10)
plt.ylabel('营业额', fontsize = 10)

plt.plot(month, sales, label = '产品A', color = 'pink')
#图例看绘画标签来确定
plt.legend(loc = 'upper left')

#网格线
# plt.grid(True) x和y轴都有
plt.grid(axis = 'y', alpha = 0.5, color = 'pink', linestyle = '--')  #alpha透明度 linestyle线的类型

#设置字体大小
plt.xticks(rotation = 30, fontsize = 27)
plt.yticks(rotation = 0, fontsize = 10)

#设置y轴的范围
plt.ylim(80, 220)

for x, y in zip(month, sales):
    print(x, y)
    plt.text(x, y + 2, str(y), ha = 'center', va = 'bottom', fontsize = 10)

plt.show()
```

柱状图

```python

#画布
plt.figure(figsize = (10, 5))
subject = ['语文', '数学', '英语', '科学']
score = [95, 120, 80, 110]

#绘制柱状图
plt.bar(subject, score, label = '李华', color = 'green')

plt.title('李华成绩', color = 'pink', fontsize = 20)
plt.xlabel('学科', fontsize = 10)
plt.ylabel('成绩', fontsize = 10)

#图例看绘画标签来确定
plt.legend(loc = 'upper left')

#网格线
# plt.grid(True) x和y轴都有
plt.grid(axis = 'y', alpha = 0.5, color = 'pink', linestyle = '--')  #alpha透明度 linestyle线的类型

#设置字体大小
plt.xticks(rotation = 30, fontsize = 27)
plt.yticks(rotation = 0, fontsize = 10)

#设置y轴的范围
plt.ylim(0, 150)

for x, y in zip(subject, score):
    print(x, y)
    plt.text(x, y + 2, str(y), ha = 'center', va = 'bottom', fontsize = 10)

#自动优化排版
plt.tight_layout()

plt.show()
```

环形图 + 爆炸

```python

#画布
plt.figure(figsize = (10, 5))
things = ['吃饭', '睡觉觉', '打豆豆', '划水']
times = [2, 10, 4, 8]

color = ['pink', '#66b3ff', 'yellow', '#ff9999']
explode = [0, 0.1, 0, 0] #设置突出块的位置

plt.pie(times, labels = things, autopct='%.1f%%', startangle=90, 
        colors = color, wedgeprops = {'width': 0.6}, pctdistance = 0.7,
        explode = explode, shadow = True) 
#wedgeprops设置圆环宽度 pctdistance设置百分比的位置
plt.text(0, 0, '总计100%', ha = 'center', va = 'bottom')
# ha = 'center', va = 'bottom'居中

plt.title('李华时间分布', color = 'pink', fontsize = 20)
# plt.xlabel('GDP', fontsize = 10)
# plt.ylabel('国家', fontsize = 10)

#图例看绘画标签来确定
# plt.legend(loc = 'upper left')

#网格线
# plt.grid(True) x和y轴都有
# plt.grid(axis = 'x', alpha = 0.5, color = 'pink', linestyle = '--')  #alpha透明度 linestyle线的类型

#设置字体大小
# plt.xticks(rotation = 30, fontsize = 27)
# plt.yticks(rotation = 0, fontsize = 10)

#设置y轴的范围
# plt.ylim(0, 150)

# for x, y in zip(subject, score):
#     print(x, y)
#     plt.text(x, y + 2, str(y), ha = 'center', va = 'bottom', fontsize = 10)

#自动优化排版
plt.tight_layout()

plt.show()
```
