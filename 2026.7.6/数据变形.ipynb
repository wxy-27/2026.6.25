```python
import pandas as pd
data = {
    'ID': [1, 2], 
    'name': ['A', 'B'],
    'English': [60, 70],
    'Math': [70, 60],
    'Science': [70, 80]
}
df = pd.DataFrame(data)
df.T
```

```python
df2 = pd.melt(df, id_vars = ['ID', 'name'], var_name = '科目', value_name = '分数')  #宽表转换成长表
```

```python
df2.sort_values('name')  #按名字排序
```

```python
pd.pivot(df2, index=['ID', 'name'], columns = '科目', values = '分数') #长转宽
```
