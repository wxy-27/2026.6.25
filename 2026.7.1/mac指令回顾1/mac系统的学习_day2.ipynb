
**文件夹与文件不能弄乱**


如果你想要在 Mac 的终端（Terminal）里游刃有余，学习 Unix 命令确实是一把开启新世界大门的钥匙。Mac 的 macOS 底层是基于 BSD Unix 的，因此绝大多数 Linux/Unix 命令在这里都通用。
不需要一口气吃成个胖子，我们可以把要学的命令分为**五个核心阶段**：
## 阶段一：基础导航与文件管理（最核心）
这是你每天用得最多的命令，用来替代鼠标在 Finder（访达）里的点击操作。
 * pwd (Print Working Directory)：查看你当前在哪个文件夹。
 * ls：列出当前目录下的文件。       **（查看当前目录下的所有文件）**
   * ls -l：显示详细信息（权限、大小、修改时间）。
   * ls -a：显示所有文件（包括以 . 开头的隐藏文件）。
 * cd (Change Directory)：切换文件夹。              **cd name1 （在当前目录下访问其他已有文件夹）**
   * cd ..：返回上一级。               **中间有空格**
   * cd ~：直接回到家目录（User 目录）。        **中间有空格**
 * mkdir：新建文件夹。       **mkdir name1(新建名称)**
 * touch：新建空白文件。        **touch name1(新建名称)**
 * cp：复制文件或文件夹（复制文件夹用 cp -r）。      **cp name1(已有) name2(新建名称)**
 * mv：移动文件（也用来重命名文件）。         **cp name1(已有) name2(name1 改名为name2)**
 * rm：删除文件。**（注意：Unix 的删除不进废纸篓，直接消失！）**
   * rm -rf：强制删除文件夹及其内容（威力巨大，使用时要格外小心）。      **rm -r name1(删除文件夹要-rf 文件不用)**
## 阶段二：文件内容查看与编辑
不用打开第三方软件，直接在终端里读写文件。
 * cat：一次性在终端打印出整个文件的内容。           **cat name1(查看文件内容)  cat -n name1(有id标注第几行)**
 * less：分屏查看大文件（按空格翻页，按 q 退出）。        **(看的是文件内容 不过想是把它打开了来看)**
 * head / tail：查看文件的前几行或后几行 **（tail -f 常用来看实时动态日志  ？（就可以编辑嘛 似乎也改不了文件内容））。         (看文件内容默认10行)              tail / head -n 5 name1(已有文件名称)  查看后 / 前 5行信息**
 * nano：极其适合新手的终端文本编辑器。    **nano name1(已有文件)  编译文件内容control + o/x/w/k/u/g   保存/退出/搜索/剪切/粘贴/新人帮助指南**
 * vim：功能极其强大的编辑器（有学习曲线，但学会了就是神器）。
 * **vim name(可以是存在的文件也可以另外新建) vim +n name 可以直接定位在n行 不加n会到底部 i进入编辑 Esc退出编辑 ':q!'会退出编辑  ':w'保存并退出   ':w name'(文件名)另外新建在外面       不太熟悉很多关于vim的相关指令**
## 阶段三：权限管理与系统信息

**如果是被sudo修饰的命令出来的文件夹执行里头的子目录都要'sudo 命令'**

macOS 对安全性要求很高，经常需要用到权限命令。
 * sudo (SuperUser Do)：以管理员权限执行命令（会要求输入 Mac 开机密码）。  **就是在指令前加sudo做好后想打开用指令做的要输入开机密码**
 * chmod：修改文件或文件夹的权限（比如让一个脚本变成“可执行”：chmod +x script.sh）。
 * **设置文件为 755 权限​​：chmod 755 name（所有者读写执行，其他人读执行）设置文件为 777 权限：chmod 777 name（所有用户完全控制，慎用）    递归修改目录权限：chmod -R 755 folder_name    移除其他用户的写权限：chmod go-w name
   只允许所有者读写：chmod 600 filename**
 * chown：修改文件所属的用户或用户组             **将文件交给管理员组： sudo chown :admin name1     将文件交给用户组：chown :staff name1                   //  whoami --> 查看当前用户名                      groups  --> 查看当前用户在哪些组(?)**
 * top 或 htop：查看系统进程和内存/CPU 占用情况（相当于命令行版的活动监视器）。  **给一团东西看不懂 按q退出 其中可以专门监控 -stats cpu/vm(cpu/内存)**
 * ps -ef：列出当前运行的所有进程。**给一团东西看不懂**
 * kill：结束某个进程（配合进程 ID，如 kill -9 1234 强制结束）**懵 似乎是关闭进程（可以靠ps -ef|grep ping来看那些进程正在运行并且4位数字位进程名称）实际操作不太会**  
## 阶段四：网络与文本处理（进阶）

当你需要调试网络或者批量处理文本数据时。
 * ping：测试网络连通性。     **给一团东西看不懂  测试到百度的连通性​​：ping baidu.com  ping也可以算开启进程
​​只 ping 5 次就退出​​：ping -c 5 URL
​​快速检查本地网络​​：ping -c 3 192.168.1.1（路由器地址）
​​每 0.5 秒 ping 一次​​：ping -i 0.5 URL
​​静默模式看结果​​：ping -c 10 -q URL
​​测试 IPv6 连接​​：ping6 URL**
 * curl：下载文件或发送网络请求（命令行版的浏览器/Postman）。**给一团东西看不懂 ​​获取网页内容​​：curl URL
​​查看响应头信息​​：curl -I URL
​​下载文件到本地​​：curl -O URL.zip
​​带进度条下载​​：curl --progress-bar -O URL.zip
​​发送 POST 数据​​：curl -X POST -d "name=test" URL
​​携带 JSON 数据​​：curl -X POST -H "Content-Type: application/json" -d '{"key":"value"}' URL
​​上传文件​​：curl -F "file=@~/Desktop/test.txt" URL
​​跟随重定向​​：curl -L URL**
 * grep：极其强大的文本搜索工具（比如 ls | grep "photo" 可以在文件列表中筛选出包含 photo 的文件）。
**类似文本查找功能 grep AA name （在name文件查找AA包含AA一行全部返回会区分大小写）
如果不会区分大小写 AA前加个-i 如果想精确匹配就加-w 如果想查找多个就加-e（并集）如果确定行数加-n 如果想要反选-v 如果想在文件里找-r**
 * awk / sed：文本流处理“三剑客”的另外两位，适合批量修改文本（前期了解即可）。
## 阶段五：Mac 专属特有命令
这些是 macOS 系统特有的命令，Linux 里可没有哦！
 * open：用 Mac 的默认程序打开文件或文件夹（如 open . 会直接在访达中打开当前目录）。 **懂之   open 文件/文件名/应用程序名称 前提是在当前目录内的**
 * pbcopy 和 pbpaste：**超级好用！** 与 Mac 的剪贴板互动。  **pbpaste粘贴功能  pbpaste|less可以看粘贴的内容 pbcopy不会用**
   * cat file.txt | pbcopy（把文件内容复制到剪贴板）
   * pbpaste > new.txt（把剪贴板内容写入新文件）
 * softwareupdate：命令行版的系统更新。
 * **检查可用更新：sudo softwareupdate -l
安装所有更新：sudo softwareupdate -i -a
安装指定更新：sudo softwareupdate -i [更新名称]
重启后安装：sudo softwareupdate -R**
 * screencapture：命令行截屏。 **screencapture ~/Desktop/123.png (放在桌面上的截图名为123.png)**
### 💡 给新手的两个保命/进阶技巧
 1. **Tab 键自动补全**：打出路径或命令的前几个字母，狂按 Tab 键，系统会自动帮你补全，既省时又不会出错。
 2. **man 命令（男人灯塔）**：任何命令不知道怎么用，在前面加 man。比如 man ls，系统就会把 ls 的完整官方使用说明书展示给你（按 q 退出）。
**建议学习路径：** 先花半天时间把“阶段一”的命令在终端里自己敲几遍，熟悉在没有鼠标的情况下怎么在文件夹之间“反复横跳”。当你习惯了之后，你会发现命令行有时候比图形界面高效得多！
