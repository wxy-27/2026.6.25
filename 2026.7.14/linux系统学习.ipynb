# Linux系统指令学习

## echo指令

1：变量=值            可以赋值  
2：echo 'string'     输出字符串（可以不带单引号）  
3：echo "$变量"      输出变量值    

4: echo $?          如果出非0说明上条语句有问题  
5: echo -n string   -n去回撤（默认自带回撤）  
6：echo \ ...        续行符'\'回撤后可以继续写（用于太长语句）  
7：!!                有n个!就调用上n个命令  
8：!字符             向上搜倒数第一个字符命令 显示并执行  
9：clear              清屏  
10: echo -e " string"  在string里把\转义   \n \t ...  
11: echo -e ""       可以在引号里定义字符串颜色\e(**不过我没实现出来**)

## cd指令  
1:pwd              当前目录  
2：cd /Desktop    去桌面目录  
3: cd ~           回到最初目录  
4: cd ..         回到上一级目录  
5: cd -          回到上次访问的目录  
6: cd /          回到根目录  
7: ll            访问当前目录下的文件/ **ls为macOS的指令文件**

## find指令  
**tip：macOS会有很多查询的进程会显示**  
1: find . -name "文件名"       查找文件  
2：find . -name "*.txt"        查找所有.txt格式的文件  
3：find . -iname "文件名"      不区分大小写查找文件  
4：find . -type 字符           查找同类型的..文件  
5：find . -mmin +n             在n分钟之前编辑出了什么文件  
6：find . -mmin -n             在n分钟之内编辑出了什么文件  
7：find . -mtime +n             在n天之前编辑出了什么文件  
8：find . -mtime -n             在n天之内编辑出了什么文件     
9：find . -user user名           查找用户  **（还未实现）**  
10：find . -group group名        查找组别  **（还未实现）**

## mkdir指令  
1：mkdir text1              创建text1文件夹  
2：mkdir -p text1/text2     在text1文件夹里创建text2子文件夹  
3：mkdir -vp text1/text2    在text1文件夹里创建text2子文件夹（创建流程会打印出来）  
4：mkdir -m 700 text1       只有当前用户可以执行读写的权限  

## cp指令
1：cp A文件 B文件      在当前目录下的文件A复制一份为B且存储在当前目录  
2：cp -v A B    打印执行过程  
3：cp -r A文件夹 B文件夹   在当前目录下的文件A复制一份为B且存储在当前目录  
4：cp -i A文件  B文件（存在）      询问是否覆盖B（无-i 默认覆盖）  

## cat指令
1：cat A文件          查看A文件内容  
2：cat -n A文件       查看A文件内容并且标系统记每一行行数  
3：cat -b A文件       查看A文件内容并且标记文本每一行行数（空行会正常显示）  
4：cat -s A文件       查看A文件内容并且标记文本每一行行数（空行不会过多显示 多空行会合并成一行） 
5：cat -E  A文件      在每个自主换行的位置加个$  **(tip:macOS系统是-e)**  
6：cat -T  A文件      制表符(多个也只会有一个^)用^表示  **(tip：macOS系统是-t)**  
7：cat -A  A文件      同时实现-E和-T功能              **（还未实现 报错）**  
8：cat A文件 B文件 .. 可以同时查多个文件内容  
9：cat 路径名          访问路径里的信息内容  

## rm指令  
1：rm A文件  删除A文件  
2：rm -r A文件夹 删除A文件夹  
3：rm -i  A文件  询问你是否删除A文件 **（tip：-i可以叠用 位置可以换来换去 -r -i/-i -r）**

## wc指令
1：wc A文件 描述A文件  出现3个参数 分别对应行数、单词数、字节数  
2：wc -l  A文件       文件行数  
3：wc -w  A文件       文件单词数  
4：wc -c  A文件       文件字节数（1汉字等于2字节）  
5：wc -m  A文件       文件字符数  

## tar指令
**tip：当前参数为组合形式 注意f一定要放后面 其他没强制要求**  
**-c压缩 -x解压 -z为gz压缩 -v描述 -f文件名**   
  
1：tar -cf A.tar B文件夹           A:创造一个压缩文件（可以重新命名）B：压缩文件夹的名称  
2：tar -xf A.tar                  解压A（名称还是压缩前名称）
3：tar -zcf A.tar.gz B文件夹      重新把B压缩为.tar.gz  
4：tar -zxvf A.tar.gz             描述A详细的解压过程

## tail/head指令  

1：tail/head A文件     打印末尾/开头10行数据（默认）  
2：tail/head -n 50 A文件    打印末尾/开头min(50, 数据总长)行数据  
3：tail -n +50 A文件   从第50行开始打印数据  
4：head -n -50 A文件   从倒数第50行开始打印数据  
5：tail/head -f A文件       实时监视 **（还未实现，感觉有点抽象）**  
6：tail/head   同5，但是文件丢失后会等待文件新建**（还未实现，感觉有点抽象）**  

## vi指令
**感觉和vim很像**  
**i进入编辑 Esc退出编辑 ：底线命令**  

i 在当前字符前插入文本  
I 在行首插入文本  
a 在当前字符后添加文本  
A 在行末添加文本  
o 在当前行后面插入一空行  
O 在当前行前面插入一空行  
^ 行首，第一个不是空白字符的位置  
$ 行尾  
gg 文件顶部  
G 文件末尾  
数字gg 移动到 数字 对应行数  
数字G 移动到 数字 对应行数  
u 撤销上次命令  
U 撤销当前用户的所有的编辑操作  
Ctrl+r 恢复撤销的命令  
Ctrl+b 向上翻页  
Ctrl+f 向下翻页  
Ctrl+g 显示光标所在位置的行号和文件的总行数  
v 从光标位置开始按照正常模式选择文本  
V 选中光标经过的完整行  
x 删除光标所在字符，或者选中文字  
dd 删除光标所在行，可以 ndd 删除多行  
D 删除至行尾  
yy 复制一行，可以 nyy 复制多行  
p 粘贴  
/str 查找 str  
n 查找下一个(必须在查找状态下执行)  
N 查找上一个(必须在查找状态下执行)  
gUU 将当前行的字母全改成大写  
guu 将当前行的字母全改成小写  
gUw 将光标下的单词改成大写  
guw 将光标下的单词改成小写  
g~~ 行翻转(大小写互换)  
ZZ 保存并退出  
  
ZQ 不保存并退出# 末行模式(:):  
:数字 移动到 数字 对应行数  
:%d 删除全部  
:%s/旧文本/新文本/g 查找并替换    
:%s/旧文本/新文本/gc 查找并替换，会有提示(y-替换；n-不替换；q-退出替换；l-最后一个，并把光标移动到行首)  
:r 文件的路径及文件名 在当前文件中读入其他文件的内容  
:r! 命令名 将某一命令的结果追加到当前行的下一行  
:w 保存  
:w 文件的路径及文件名 另存为其他文件  
:q 退出，如果没有保存，不允许退出   
:w! 强行保存  
:q! 强行退出，不保存退出  
:wq 保存并退出  
:wq! 强制保存并退出  
:x 保存并退出  
:set nu 显示行号  
:history 列出历史命令记录  
:pwd 显示当前所在的路径
