预测高阶

```python
from ultralytics import YOLO

# 加载模型
model = YOLO("yolo26n.pt")  # 加载官方预训练模型
# model = YOLO("path/to/best.pt")  # 加载自定义训练好的模型

# 使用模型进行预测
results = model("https://ultralytics.com/images/bus.jpg")  # 对图片执行目标检测预测

# 解析预测结果
for result in results:
    xywh = result.boxes.xywh  # 边界框坐标：中心x、中心y、框宽度、框高度
    xywhn = result.boxes.xywhn  # 归一化后的xywh坐标（相对图片尺寸0~1）
    xyxy = result.boxes.xyxy  # 边界框坐标：左上角x、左上角y、右下角x、右下角y
    xyxyn = result.boxes.xyxyn  # 归一化后的xyxy坐标（相对图片尺寸0~1）
    names = [result.names[cls.item()] for cls in result.boxes.cls.int()]  # 每个检测框对应的类别名称
    confs = result.boxes.conf  # 每个检测框的置信度分数
```

**姿势评估**

```python
from ultralytics import YOLO

# 加载人体关键点姿态模型
model = YOLO("yolo26n-pose.pt")

# 必须指定coco8-pose小型姿态数据集，避免路径报错
metrics = model.val(data="coco8-pose.yaml")

# 检测框指标
print("框mAP50-95：", metrics.box.map)
print("框mAP50：", metrics.box.map50)
print("框mAP75：", metrics.box.map75)

# 关键点姿态指标
print("关键点mAP50-95：", metrics.pose.map)
print("关键点mAP50：", metrics.pose.map50)
print("关键点mAP75：", metrics.pose.map75)
```

**obb (可以画斜着的框框)**

```python
#验证

from ultralytics import YOLO

# 加载模型
model = YOLO("yolo26n-obb.pt")  # 加载官方预训练旋转框检测模型
# model = YOLO("path/to/best.pt")  # 加载自己训练完成的自定义旋转框模型

# 在验证集评估模型精度
metrics = model.val(data="dota8.yaml")
# 说明：指定dota8小型航空影像数据集；若此前执行过train训练缓存数据集，可省略data参数

# 旋转框OBB精度指标
metrics.box.map        # 旋转框综合mAP（IOU阈值0.5~0.95均值）
metrics.box.map50      # IOU=0.5时旋转框mAP
metrics.box.map75      # IOU=0.75时旋转框mAP
metrics.box.maps       # 列表，存储每一类目标各自的mAP50-95值
metrics.box.image_metrics  # 字典，按单图存储指标：精确率、召回率、F1分数、真正例、假正例、假负例
```
