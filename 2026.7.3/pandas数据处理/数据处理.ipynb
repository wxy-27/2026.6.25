**数值的缺失**

```python
import pandas as pd
import numpy as np

s = pd.Series([1, 2, np.nan, None, pd.NA])
# print(s)
print(s.isna())
print(s.isnull())
```

```python
df = pd.DataFrame([[1, pd.NA, 2], [2, 3, 4], [None, 4, 5]], columns = ['第一列', '第二列', '第三列'], index = [1, 2, 3])
print(df.isna())
```

```python
#查看缺失值数量
print(df.isna().sum())  
print(df.isna().sum(axis = 1))  
```

**剔除缺失值 dropna（）**

```python
print(s.dropna())
```

```python
print(df.dropna())  #有缺失值就剔除
```

```python
print(df.dropna(how = 'all'))  #所有都缺失才剔除
```

```python
print(df.dropna(thresh = 2))  #至少有n个数据保留
```

```python
print(df.dropna(axis = 1))
```

```python
print(df.dropna(subset = ['第二列'])) #保留第二列存在的值
```

**缺失值的填充**

```python
df = pd.read_csv('users_data.csv')
df.isna().sum(axis = 0)
print(df.head())
```

```python
print(df.fillna({'Score': 88.8, 'City': 'Hangzhou'})) #把Score和City里缺失的数值都变成88.8和Hangzhou
```

```python
print(df.fillna(df[['Age', 'Score']].mean()))  #使用平均值去填充NaN（统计值）
```

```python
print(df.ffill()) #front用前面的相邻值填充
```

```python
print(df.bfill()) #back用后面的相邻值填充
```

**数据的重复**

```python
# import pandas as pd
data = {
    "name": ['A', 'B', 'A', 'D', 'A', 'B'],
    "age": [1, 2, 3, 4, 1, 2],
    "city": ['Y', 'Y', 'S', 'S', 'Y', 'Y']
}
ss = pd.DataFrame(data)
ss
```

```python
ss.duplicated()  #检查重复行
```

```python
ss.drop_duplicates() #去除重复的数据
```

```python
ss.drop_duplicates(subset = ['name']) #去除name重复的数据
```
