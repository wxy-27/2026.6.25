```python
s = pd.Series([1, 2, 3, 4], index = ["A", "B", "C", "D"])
s.name = "test"
print(s)
print(s.index)
print(s.values)
print(s.shape, s.ndim, s.size)
print(s.dtype, s.name)

# 显示索引
print(s.loc['A':'D']) 

# 隐式索引
print(s.iloc[0:2])  

# 这个不允许切片
print(s.at['A']) 
```

```python
s = pd.Series([1, 2, 3, 4], index = ["A", "B", "C", "D"])
print(s["A"])
print(s[s < 3])
# 默认五行数据
print(s.head(2))
print(s.tail(1))
```

```python
import pandas as pd
import numpy as np
s = pd.Series([1, 2, np.nan, np.nan], index = ['A', 'B', 'C', 'D'], name = 'nonono')
# 描述信息
print(s.describe())

# 约等于 index：属性 keys（）：方法
print(s.keys())

# 寻找缺失值
print(s.isna())

# 查看是否存在
print(s.isin([1, 3]))

print(s.std())

print(s.var())
```
