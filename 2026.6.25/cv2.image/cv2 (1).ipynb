```python
import pandas as pd
import cv2
import numpy as np
import time
from matplotlib import pyplot as plt
img = cv2.imread("dog.jpg")
plt.imshow(img[:,:,::-1])
a = np.array(img)
```

```python
b = img.copy()
h = len(a)
w = len(a[0])
st = time.time()
for i in range(h):
    for j in range(w):
        for k in range(3):
            b[i][j][k] = min(int(b[i][j][k]) + 20, 255)
ed = time.time()
print(ed - st)
plt.imshow(b[:,:,::-1])
```
